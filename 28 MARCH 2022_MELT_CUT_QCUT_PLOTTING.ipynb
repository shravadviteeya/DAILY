############### CUT and QCUT FUNCTION ############################
CUT FUNCTION DIVIDES THE VALUES IN SPECIFIC RANGE
DATA BINNING: NO OF BINS IN WHICH DATA TO BE PARTITIONED 
Bin labels must be one fewer than the number of bin edges


```python
df3 = pd.read_csv("G:\\My Drive\\Data\\StudentsPerformance.csv")
df3
```

```python
df3.columns
```

```python
df4 = df3.drop(columns= ['race/ethnicity', 'parental level of education', 'lunch',
       'test preparation course'])
```

```python
df4.head()
```

```python
pd.cut(df4["math score"], bins = 4)
```

```python
pd.cut(df4["math score"], bins = 4).value_counts()
```

```python
pd.cut(df4["math score"], bins = 5)
```

```python
pd.cut(df4["math score"], bins = 5).value_counts()
```

```python
grades =  ['C','B-','B+','A']  # bin lables
cut_bins = [0,  20, 50, 80, 100   ]
```

```python
df4["grades"] = pd.cut(df4["math score"], bins = cut_bins )
df4
```

```python
df4["grades"] = pd.cut(df4["math score"], bins = cut_bins , labels= grades)
df4
```

```python
############################# QCUT###############################
#(-0.001, 57.0]    266 here it excludes -0.001 and starts from 0 to 57 
#(66.0, 77.0]      264 here it excludes 66 and starts from 67 to 77 
#(57.0, 66.0]      241 here it excludes 57 and starts from 58 to 66
#(77.0, 100.0]     229 here it excludes 77 and starts from 78 to 100
```

```python
pd.qcut(df4["math score"], q=4)
```

```python
pd.qcut(df4["math score"], q=4).value_counts()
```

```python
# q cut comes from the describe function 
```

```python
df4['math score'].describe()
```

```python
pd.qcut(df4["math score"], q=6).value_counts()
```

```python
df4["math_q_labels"] = pd.qcut(df4['math score'], q = [0,  .20, .50, .80, 1] , labels= grades)
```

```python
df4
```

```python
df4["math_q_labels_false"] = pd.qcut(df4['math score'], q = [0,  .20, .50, .80, 1] , labels= False)
df4
```

```python
#   q = [0,  .20, .50, .80, 1]   = math_q_labels_false [ 0 , 1 , 2 , 3 , 4]
```

```python
# how to know the score 
```

```python
results , bin_edges = pd.qcut(df4['math score'], q = [0,  .20, .50, .80, 1] , labels= grades , retbins= True)
```

```python
results
```

```python
bin_edges
```

```python
result_table = pd.DataFrame(zip(bin_edges , grades) ,columns= ['Limits ' , 'grades'])
result_table
```
